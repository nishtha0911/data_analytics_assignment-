```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('company_sales_data.csv')
data
```

```python
sns.barplot(x=data['facecream'],y=data['month_number'],data=data)
plt.title('facecream sales')
plt.xlabel('facecream')
plt.ylabel('moth_number')
plt.show()
```

```python
sns.barplot(x=data['facewash'],y=data['month_number'],data=data)
plt.title('facewash sales')
plt.xlabel('facewash')
plt.ylabel('moth_number')
plt.show()
```

```python
sns.histplot(y=data['total_profit'],x=data['month_number'],data=data)
plt.title('total _profit corresponing to months')
plt.ylabel('TOTAL_PROFIT')
plt.xlabel('moth_number')
plt.show()
```
