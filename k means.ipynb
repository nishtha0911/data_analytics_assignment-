```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
data1 = pd.read_csv('content.csv')
```

```python
x= data1['Duration']
y = data1['Calories']
```

```python
data = list(zip(x,y))
print(data)
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(data)
```

```python
plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
